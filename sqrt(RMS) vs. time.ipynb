# Plot sqrt(<A^2>) vs. Time 
# Prediction: Linear growth
## Data pulled from long 10^5 run: https://nbviewer.jupyter.org/github/ctandon11/Mode_Turbulence_Coupling/blob/master/long%20run%2010%5E5.ipynb

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
RMS = np.array([0, 1.0769361864663842, 3.998546897166391, 8.283180022180169, 3.726129841893485, 15.496552639339006, 53.551829362977415, 125.60771106056062, 368.165833391085, 369.74345352796405, 350.9929008457846])
Time = np.array([0, 5000, 15000, 25000, 35000, 45000, 55000, 65000, 75000, 85000, 95000 ])
```

```python
sqrtRMS = []
for i in RMS:
    sqrtRMS.append(i**(1/2))
```

```python
m, b = np.polyfit(Time, sqrtRMS, 1)

plt.plot(Time, sqrtRMS, label = "Data")
plt.plot(Time, m * Time + b, label = "Line of Best Fit")
plt.xlabel("Time")
plt.ylabel("sqrt<A^2>")
plt.legend()
```
